```python
import tensorflow as tf
import numpy as np
import cv2
import os
```

```python
%cd '/content/drive/MyDrive/signlanguage'
```

```python
# Load the trained model
model = tf.keras.models.load_model("inceptionv3_model1.h5")
```

```python
# Path to test image
TEST_FOLDER = "test"
test_image_path = os.path.join(TEST_FOLDER, os.listdir(TEST_FOLDER)[0])  # Get the first image
print(test_image_path)

# Load and preprocess image using OpenCV
img = cv2.imread(test_image_path)  # Read image
img = cv2.resize(img, (299, 299))  # Resize for InceptionV3
img = img.astype("float32") / 255.0  # Normalize
img = np.expand_dims(img, axis=0)  # Add batch dimension
```

```python
class_labels =['A', 'B', 'C', 'D']
```

```python
predictions = model.predict(img)
predicted_class_index = np.argmax(predictions)  # Get index of highest probability class

predicted_class = class_labels[predicted_class_index]
print("Predicted Class: ", predicted_class)
```
